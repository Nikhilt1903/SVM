# Task 2: Data Preprocessing


```python
#Encode categorical variables if necessary.
#we are label encoding the data as there are multiple fetures

from sklearn.preprocessing import LabelEncoder

#Initializing LabelEncoder
label_encoder = LabelEncoder()

for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = label_encoder.fit_transform(data[column])
```

```python
data.head()
```

```python
#all the data is now encoded
```

```python
#Split the dataset into training and testing sets.
#Features and target variable
x = data.drop(columns='class')
y = data['class']
```

```python
from sklearn.model_selection import train_test_split

#Splitting the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

# Task 3: Data Visualization

```python
# Scatter plot between two continuous variables
sns.scatterplot(x='stalk_height', y='cap_diameter', data=data)
plt.title('Scatter Plot: Stalk Height vs Cap Diameter')
plt.xlabel('Stalk Height')
plt.ylabel('Cap Diameter')
plt.show()
```

```python
#Visualize class distributions to gauge dataset balance or imbalance.
import seaborn as sns
import matplotlib.pyplot as plt

#Plotting class distribution
sns.countplot(data=data, x='class', palette='Set1')
plt.title("Mushroom Class Distribution: Edible vs Poisonous")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

print(data['class'].value_counts())
```

```python
#this cleary shows imbalance in the dataset
```

```python
#we are balaning =the imbalance using smote
from imblearn.over_sampling import SMOTE
#Applying smote only on training data
smote = SMOTE(random_state=42)
x_train_bal, y_train_bal = smote.fit_resample(x_train, y_train)

#Confirm the new class distribution
from collections import Counter
print("Before SMOTE:", Counter(y_train))
print("After SMOTE:", Counter(y_train_bal))
```

```python
import pandas as pd
sns.countplot(x=pd.Series(y_train_bal), palette='Set2')
plt.title("Balanced Class Distribution After SMOTE")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
#the data is balanced
```

# Task 4: SVM Implementation


```python
from sklearn.svm import SVC
#Initializing SVM classifier
svm_model = SVC(kernel='rbf', random_state=42)
#Train the model
svm_model.fit(x_train_bal, y_train_bal)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
#Making predictions
y_pred = svm_model.predict(x_test)

#Evaluating performance
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1:.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# Task 6: Parameter Tuning and Optimization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_bal_scaled = scaler.fit_transform(x_train_bal)
X_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform
from sklearn.svm import SVC

# Define a smaller, faster hyperparameter space
param_dist = {
    'kernel': ['linear', 'rbf'],
    'C': uniform(0.1, 10),              # Randomly sample C from 0.1 to 10
    'gamma': ['scale', 0.01, 0.1, 1]    # Only applies to 'rbf'
}

# Setting up RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=SVC(),
    param_distributions=param_dist,
    n_iter=10,                  # Only try 10 random combinations
    scoring='accuracy',
    cv=3,                       # 3-fold CV for speed
    random_state=42,
    n_jobs=-1,
    verbose=1
)

# Fit to training data
random_search.fit(X_train_bal_scaled, y_train_bal)
```

```python
# Best parameters and score
print("Best Parameters:", random_search.best_params_)
print("Best Accuracy:", random_search.best_score_)

#Using the best model
best_svm = random_search.best_estimator_

#Predicting on test set
y_pred_opt = best_svm.predict(X_test_scaled)

#Evaluating
from sklearn.metrics import classification_report

print("\nClassification Report (Optimized SVM):")
print(classification_report(y_test, y_pred_opt))
```

```python
#our accuracy of the model improved afetr parameter Tuning and Optimization
```

# Task 7: Comparison and Analysis

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train_bal)
X_test_scaled = scaler.transform(x_test)

#Define and evaluate models with different kernels
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
results = {}

for kernel in kernels:
    print(f"\nTraining SVM with {kernel} kernel...")
    model = SVC(kernel=kernel)
    model.fit(X_train_scaled, y_train_bal)
    y_pred = model.predict(X_test_scaled)

    # Evaluate
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    
    results[kernel] = {
        'accuracy': accuracy,
        'precision': report['weighted avg']['precision'],
        'recall': report['weighted avg']['recall'],
        'f1_score': report['weighted avg']['f1-score'],
        'report': classification_report(y_test, y_pred)
    }

#Displaying Results
print("\n SVM Kernel Comparison:")
print("{:<10} {:<10} {:<10} {:<10} {:<10}".format("Kernel", "Accuracy", "Precision", "Recall", "F1 Score"))
for kernel, metrics in results.items():
    print("{:<10} {:.4f}     {:.4f}     {:.4f}   {:.4f}".format(
        kernel, metrics['accuracy'], metrics['precision'], metrics['recall'], metrics['f1_score']))
```
